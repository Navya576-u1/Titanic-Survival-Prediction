```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('Documents/Titanic.csv')
df
```

```python
df=df.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode())
```

```python
from sklearn.preprocessing import LabelEncoder
l_sex=LabelEncoder()
l_emb=LabelEncoder()
df['Sex']=l_sex.fit_transform(df['Sex'])
df['Embarked']=l_emb.fit_transform(df['Embarked'])
```

```python
l_sex.transform(['male'])
```

```python
df
```

```python
x=df.drop('Survived',axis=1)
y=df['Survived']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
import pickle
pickle.dump(model,open('model.pkl','wb')) #save model
pickle.dump(l_sex,open('l_sex.pkl','wb'))#save encodechyth oky save chyth vekanm
pickle.dump(l_emb,open('l_emb.pkl','wb'))

```

```python
model=pickle.load(open('model.pkl','rb'))
l_sex=pickle.load(open('l_sex.pkl','rb'))
l_emb=pickle.load(open('l_emb.pkl','rb'))
```
